### Solution 1: Recursion

#### code

```python
class Solution:
    def fib(self, N):
        if N == 0 or N == 1:
            return N
        return self.fib(N - 1) + self.fib(N - 2)
```

#### test

```python
N = 4
```

```python
s = Solution()
```

```python
s.fib(4)
```

### Solution 2: Dynamical Programming

#### code

```python
class Solution:
    def fib(self, N):
        f = 0 
        g = 1
        for i in range(N):
            g += f
            f = g - f
        return f
```

#### test

```python
s = Solution()
```

```python
s.fib(4)
```
