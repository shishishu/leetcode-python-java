### Solution 1: binary search

#### code

```python
class Solution:
    def singleNonDuplicate(self, nums):
        low = 0
        high = len(nums) - 1
        while low < high:
            mid = (low + high) // 2
            if mid % 2 == 1:
                mid -= 1
            if nums[mid] == nums[mid+1]:
                low = mid + 2
            else:
                high = mid
        return nums[low]
```

#### test

```python
s = Solution()
```

```python
s.singleNonDuplicate([1,1,3,3,4,8,8])
```

### Solution 2: variation

#### code

```python
class Solution:
    def singleNonDuplicate(self, nums):
        low = 0
        high = len(nums) - 1
        while low < high:
            mid = (low + high) // 2
            if nums[mid] == nums[mid ^ 1]:
                low = mid + 1
            else:
                high = mid
        return nums[low]
```

```python
s = Solution()
```

```python
s.singleNonDuplicate([1,1,3,3,4,4,8])
```
