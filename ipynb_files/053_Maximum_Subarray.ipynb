### Solution 1: Brute Force

#### code

```python
class Solution:
    def maxSubArray(self, nums):
        length = len(nums)
        max_val = -2147483648
        for i in range(length):
            output = 0
            for j in range(i, length):
                output += nums[j]
                if output > max_val:
                    max_val = output
        return max_val
```

#### test

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
```

```python
s = Solution()
```

```python
s.maxSubArray(nums)
```

### Solution 2: Dynamic Programming

#### code

```python
class Solution:
    def maxSubArray(self, nums):
        length=len(nums)  
        for i in range(1,length):  
            #   
            subMaxSum=max(nums[i]+nums[i-1],nums[i])
            nums[i]=subMaxSum  # nums[i]nums
        return max(nums) 
```

#### test

```python
s = Solution()
```

```python
s.maxSubArray(nums)
```
