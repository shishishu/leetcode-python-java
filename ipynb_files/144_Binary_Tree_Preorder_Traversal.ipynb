### Solution 1: Recursive

#### code

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution:
    def preorderTraversal(self, root):
        output = []
        self.rec_preorder(root, output)
        return output
    
    def rec_preorder(self, root, output):
        if root:
            output.append(root.val)
            self.rec_preorder(root.left, output)
            self.rec_preorder(root.right, output)
```

```python
import math
def generate_tree(tree_vals):
    num_elem = len(tree_vals)
    num_level = math.log2(num_elem + 1)
    assert num_level % 1 == 0, 'wrong number in tree values...'
    # init all tree nodes
    ptree = []
    for i in range(num_elem):
        tmp_tree = TreeNode(tree_vals[i])
        ptree.append(tmp_tree)
    for i in range(int(num_elem / 2 - 1) + 1):
        ptree[i].left = ptree[2 * i + 1]
        ptree[i].right = ptree[2 * i + 2]
    return ptree[0]
```

#### test

```python
tree_vals = [1, 2, 3, 4, 5, 6, 7]
root = generate_tree(tree_vals)
```

```python
s = Solution()
```

```python
s.preorderTraversal(root)
```

### Solution 2: Iterative with Stack

#### code

```python
class Solution:
    def preorderTraversal(self, root):
        output = []
        stack = []
        if root:
            stack.append(root)
        while stack:
            x = stack.pop()
            output.append(x.val)
            if x.right:
                stack.append(x.right)
            if x.left:
                stack.append(x.left)
        return output
```

```python
s = Solution()
```

```python
s.preorderTraversal(root)
```

### Solution 3: Iterative (general)

#### code

```python
class Solution:
    def preorderTraversal(self, root):
        output = []
        stack = []
        while True:
            self.visitAlongLeftBranch(root, output, stack)
            if not stack:
                break
            root = stack.pop()
        return output
    
    def visitAlongLeftBranch(self, root, output, stack):
        while root:
            output.append(root.val)
            stack.append(root.right)
            root = root.left
```

```python
s = Solution()
```
