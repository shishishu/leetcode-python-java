### Solution 2: Dynamic Programming

#### code

```python
class Solution:
    def lengthOfLIS(self, nums):
        if not nums:
            return 0
        len_nums = len(nums)
        mem = [1] * len_nums
        result = 1
        for i in range(len_nums):
            for j in range(i):
                if nums[j] < nums[i]:
                    mem[i] = max(mem[j] + 1, mem[i])
            result = max(result, mem[i])
        return result
```
