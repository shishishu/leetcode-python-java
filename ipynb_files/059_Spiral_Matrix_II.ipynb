### Solution 1

#### code

```python
class Solution:
    def generateMatrix(self, n: int):
        output = [[0]*n for _ in range(n)]
        minRow = 0
        maxRow = n - 1
        minCol = 0
        maxCol = n - 1
        value = 0
        maxValue = n**2
        while True:
            for i in range(minCol, maxCol + 1):
                value += 1
                output[minRow][i] = value
            minRow += 1
            if value == maxValue:
                break
            for i in range(minRow, maxRow + 1):
                value += 1
                output[i][maxCol] = value
            maxCol -= 1
            if value == maxValue:
                break
            for i in range(maxCol, minCol - 1, -1):
                value += 1
                output[maxRow][i] = value
            maxRow -= 1
            if value == maxValue:
                break
            for i in range(maxRow, minRow - 1, -1):
                value += 1
                output[i][minCol] = value
            minCol += 1
            if value == maxValue:
                break
        return output
```
