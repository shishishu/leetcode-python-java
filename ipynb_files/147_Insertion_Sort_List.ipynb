### Solution 1:

#### code

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
class Solution:
    def insertionSortList(self, head):
        p = dummy = ListNode(0)
        cur = dummy.next = head
        while cur and cur.next:
            next_val = cur.next.val
            if cur.val <= next_val:
                cur = cur.next
                continue
            if p and p.next.val > next_val:
                p = dummy
            while p and p.next.val <= next_val:
                p = p.next
            new = cur.next
            cur.next = new.next
            new.next = p.next
            p.next = new
        return dummy.next
```

```python
def generate_nodelist(val_list):
    nodes = [ListNode(x) for x in val_list]
    for i in range(1, len(nodes)):
        nodes[i - 1].next = nodes[i]
    return nodes[0]
```

```python
def print_nodelist(header):
    val_list = []
    while (header):
        val_list.append(header.val)
        header = header.next
    print(val_list)
```

#### test

```python
nums = [1,3,2,5,4]
node_header = generate_nodelist(nums)
print_nodelist(node_header)
```
