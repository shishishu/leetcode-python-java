### Solution 1: Traverse

#### code

```python
class Solution:
    def rangeSumBST(self, root, L, R):
        data = []
        def inorder(root):
            if root.left:
                inorder(root.left)
            data.append(root.val)
            if root.right:
                inorder(root.right)
        inorder(root)
        return sum([x for x in data if x >= L and x <= R])
```

### Solution 2: BST features

#### code

```python
class Solution:
    def rangeSumBST(self, root, L, R):
        data = []
        def cond_inorder(root):
            if root:
                if root.val < L:
                    cond_inorder(root.right)
                elif root.val > R:
                    cond_inorder(root.left)
                else:
                    data.append(root.val)
                    cond_inorder(root.left)
                    cond_inorder(root.right)
        cond_inorder(root)
        return sum(data)
```

### Solution 3: Recursive

#### code

```python
class Solution: 
    def rangeSumBST(self, root, L, R):
        def dfs(node, L, R):
            if node:
                if node.val >= L and node.val <= R:
                    self.ans += node.val
                if node.val > L:
                    dfs(node.left, L, R)
                if node.val < R:
                    dfs(node.right, L, R)
        self.ans = 0
        dfs(root, L, R)
        return self.ans
```

### Solution 4: Iterative

```python
class Solution: 
    def rangeSumBST(self, root, L, R):
        ans = 0
        stack = [root]
        while stack:
            node = stack.pop()
            if node:
                if node.val >= L and node.val <= R:
                    ans += node.val
                if node.val > L:
                    stack.append(node.left)
                if node.val < R:
                    stack.append(node.right)
        return ans
```
