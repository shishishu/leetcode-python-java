### Solution 1: Brute Force

#### code

```python
class Solution:
    def searchInsert(self, nums, target):
        for idx, num in enumerate(nums):
            if target <= num:
                return idx
            else:
                continue
        return idx + 1
```

```python
def test_part(nums, targets, exp_results):
    s = Solution()
    for (target, exp_result) in zip(targets, exp_results):
        res = s.searchInsert(nums, target)
        assert res == exp_result, 'wrong at target: {}'.format(target)
    print('succeed...')
    return
```

#### test

```python
nums = [1, 3, 5, 6]
targets = [5, 2, 7, 0]
exp_results = [2, 1, 4, 0]
```

```python
test_part(nums, targets, exp_results)
```

### Solution 2: binary search

#### code

```python
class Solution:
    def searchInsert(self, nums, target):
        low = 0
        high = len(nums) - 1
        while (low <= high):
            middle = int((low + high) / 2)
            if nums[middle] == target:
                return middle
            elif nums[middle] < target:
                low = middle + 1
            else:
                high = middle - 1
        return low
```

#### test

```python
test_part(nums, targets, exp_results)
```

### Solution 3: one line

```python
class Solution:
    def searchInsert(self, nums, target):
        return len([x for x in nums if x < target])
```
