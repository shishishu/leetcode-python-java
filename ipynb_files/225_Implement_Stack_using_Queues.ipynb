### Solution 1

#### code

```python
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = []
        

    def push(self, x):
        """
        Push element x onto stack.
        """
        self.queue.append(x)
        

    def pop(self):
        """
        Removes the element on top of the stack and returns that element.
        """
        if self.empty():
            return
        else:
            top_element = self.queue[-1]
            self.queue = self.queue[:-1]
            return top_element
            
    def top(self):
        """
        Get the top element.
        """
        if self.empty():
            return
        else:
            return self.queue[-1]

    def empty(self):
        """
        Returns whether the stack is empty.
        """
        if len(self.queue) == 0:
            return True
        else:
            return False
```
