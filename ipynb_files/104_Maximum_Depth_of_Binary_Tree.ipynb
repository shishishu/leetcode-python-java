### Solution 1: Recursive

#### code

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution:
    def maxDepth(self, root):
        if not root:
            return 0
        return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1
```

```python
import math
def generate_tree(tree_vals):
    num_elem = len(tree_vals)
    num_level = math.log2(num_elem + 1)
    assert num_level % 1 == 0, 'wrong number in tree values...'
    # init all tree nodes
    ptree = []
    for i in range(num_elem):
        tmp_tree = TreeNode(tree_vals[i])
        ptree.append(tmp_tree)
    for i in range(int(num_elem / 2 - 1) + 1):
        ptree[i].left = ptree[2 * i + 1]
        ptree[i].right = ptree[2 * i + 2]
    return ptree[0]
```

#### test

```python
tree_vals = [3, 9, 20, None, None, 5, 17]
root = generate_tree(tree_vals)
```

```python
s = Solution()
```

```python
s.maxDepth(root)
```

### Solution 2: BFS

#### code

```python
class Solution:
    def maxDepth(self, root):
        depth = 0
        level = [root] if root else []
        while level:
            depth += 1
            queue = []
            for elem in level:
                if elem.left:
                    queue.append(elem.left)
                if elem.right:
                    queue.append(elem.right)
            level = queue
        return depth
```

### Solution 3: DFS

#### code

```python
class Solution:
    def maxDepth(self, root):
        depth = 0
        stack = [(root, 1)]
        while stack:
            root, leng = stack.pop()
            if not root:
                return 0
            if leng > depth:
                depth = leng
            if root.left:
                stack.append((root.left, leng + 1))
            if root.right:
                stack.append((root.right, leng + 1))
        return depth
```
