### Solution 1: list

#### code

```python
class MyHashSet:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.set = []
        

    def add(self, key: int):
        if not self.contains(key):
            self.set.append(key)
        else:
            pass
        
    def remove(self, key: int):
        if self.contains(key):
            self.set.remove(key)
        else:
            pass
        

    def contains(self, key: int):
        """
        Returns true if this set contains the specified element
        """
        for data in self.set:
            if key == data:
                return True
        return False
```

### Solution 2: trivial hash

```python
class MyHashSet:
    
    def __init__(self):
        self.size = 10000
        self.buckets = [[] for _ in range(self.size)]
    
    def add(self, key):
        bucket, idx = self._index(key)
        if idx >= 0:
            return
        bucket.append(key)
    
    def remove(self, key):
        bucket, idx = self._index(key)
        if idx < 0:
            return
        bucket.remove(key)
        
    def contains(self, key):
        _, idx = self._index(key)
        return idx >= 0
    
    
    def _hash(self, key):
        return key % self.size
    
    def _index(self, key):
        hash_val = self._hash(key)
        bucket = self.buckets[hash_val]
        for i, k in enumerate(bucket):
            if k == key:
                return bucket, i
        return bucket, -1
```
