### Solution

#### code

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def isSymmetric(self, root):
        def isSym(L, R):
            if not L and not R:
                return True
            if L and R and L.val == R.val:
                return isSym(L.left, R.right) and isSym(L.right, R.left)
            return False
        if not root:
            return True
        return isSym(root.left, root.right)
```

#### test

```python
import math
```

```python
def generate_tree(tree_vals):
    num_elem = len(tree_vals)
    num_level = math.log2(num_elem + 1)
    assert num_level % 1 == 0, 'wrong number in tree values...'
    # init all tree nodes
    ptree = []
    for i in range(num_elem):
        tmp_tree = TreeNode(tree_vals[i])
        ptree.append(tmp_tree)
    for i in range(int(num_elem / 2 - 1) + 1):
        ptree[i].left = ptree[2 * i + 1]
        ptree[i].right = ptree[2 * i + 2]
    return ptree[0]
```

```python
def test(tree_vals):
    root = generate_tree(tree_vals)
    s = Solution()
    return s.isSymmetric(root)
```

```python
tree_vals = [1,2,2,3,4,4,3]
test(tree_vals)
```

```python
tree_vals = [1,2,2,None,4,4,3]
test(tree_vals)
```
