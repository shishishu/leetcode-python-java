### Solution 1: DP

#### code

G(n) = F(1, n) + F(2, n) + ... + F(n, n)  # F(i, n), number i at root

G(0) = 1, G(1) = 1

F(i, n) = G(i - 1) * G(n - i) # left, right

```python
class Solution:
    def numTrees(self, n):
        if n == 0 or n == 1:
            return 1
        res = 0
        for i in range(1, n+1):
            res += self.numAtSpecPoint(i, n)
        return res
    
    def numAtSpecPoint(self, i, n):
        return self.numTrees(i - 1) * self.numTrees(n - i)
```

### Solution 2: DP+

#### code

```python
class Solution:
    def numTrees(self, n):
        res = [0] * (n + 1)
        res[0] = 1
        res[1] = 1
        for i in range(2, n + 1):
            for j in range(1, i+1):
                res[i] += res[j - 1] * res[i - j]
        return res[n]
```

### Solution 3: Recursive

#### code

```python
class Solution:
    def numTrees(self, n):
        return self.subTrees(1, n)
    
    def subTrees(self, lo, hi):
        if lo >= hi:
            return 1
        total = 0
        for i in range(lo, hi+1):
            total += self.subTrees(lo, i - 1) * self.subTrees(i + 1, hi)
        return total
```
