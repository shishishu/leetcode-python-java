### Solution 1: 

#### code

```python
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.q = []
        
    def push(self, x):
        curMin = self.getMin()
        if curMin == None or x < curMin:
            curMin = x
        self.q.append((x, curMin))
        
    def pop(self):
        if self.q:
            self.q.pop()
        else:
            raise IndexError
            
    def top(self):
        if self.q:
            return self.q[-1][0]

    def getMin(self):
        if self.q:
            return self.q[-1][1]
        else:
            return None
```
