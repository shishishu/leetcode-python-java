### Solution 1

#### code

```python
class Solution:
    def spiralOrder(self, matrix):
        if matrix == []:
            return []
        m = len(matrix)
        n = len(matrix[0])
        output = []
        minRow = 0
        maxRow = m - 1
        minCol = 0
        maxCol = n - 1
        total_cnt = m * n
        cnt = 0
        while True:
            for i in range(minCol, maxCol + 1):
                cnt += 1
                output.append(matrix[minRow][i])
            minRow += 1
            if cnt == total_cnt:
                break
            for i in range(minRow, maxRow + 1):
                cnt += 1
                output.append(matrix[i][maxCol])
            maxCol -= 1
            if cnt == total_cnt:
                break
            for i in range(maxCol, minCol - 1, -1):
                cnt += 1
                output.append(matrix[maxRow][i])
            maxRow -= 1
            if cnt == total_cnt:
                break
            for i in range(maxRow, minRow - 1, -1):
                cnt += 1
                output.append(matrix[i][minCol])
            minCol += 1
            if cnt == total_cnt:
                break
        return output
```

#### test
