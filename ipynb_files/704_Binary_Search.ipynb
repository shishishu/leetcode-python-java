### Solution 1: Binary Search

#### code

```python
class Solution:
    def search(self, nums, target):
        low = 0
        high = len(nums) - 1
        while (low <= high):
            middle = int((low + high) / 2)
            if nums[middle] == target:
                return middle
            elif nums[middle] > target:
                high = middle - 1
            else:
                low = middle + 1
        return -1
```
