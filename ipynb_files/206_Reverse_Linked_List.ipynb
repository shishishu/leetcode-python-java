### Solution 1: simple

#### code

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
class Solution:
    def reverseList(self, head):
        if not head:
            return None
        data = []
        while (head):
            data.append(head.val)
            head = head.next
        data_inv = reversed(data)
        nodes = [ListNode(x) for x in data_inv]
        for i in range(1, len(nodes)):
            nodes[i - 1].next = nodes[i]
        return nodes[0]
```

```python
def generate_nodelist(val_list):
    nodes = [ListNode(x) for x in val_list]
    for i in range(1, len(nodes)):
        nodes[i - 1].next = nodes[i]
    return nodes[0]
```

```python
def print_nodelist(header):
    val_list = []
    while (header):
        val_list.append(header.val)
        header = header.next
    print(val_list)
```

#### test

```python
nums = [1,2,3,4,5]
node_header = generate_nodelist(nums)
print_nodelist(node_header)
```

```python
s = Solution()
```

```python
head_inv = s.reverseList(node_header)
```

```python
print_nodelist(head_inv)
```

### Solution 2: Iterative

#### code

```python
class Solution:
    def reverseList(self, head):
        new_head = None
        while (head):
            tmp_node = ListNode(head.val)
            head = head.next
            tmp_node.next = new_head
            new_head = tmp_node
        return new_head
```

```python
print_nodelist(node_header)
```

```python
s = Solution()
```

```python
print_nodelist(s.reverseList(node_header))
```

### Solution 3: Recursion

#### code

```python
class Solution:
    def reverseList(self, head):
        return self._reverse(head)
    def _reverse(self, node, prev=None):
        if not node:
            return prev
        n = node.next
        node.next = prev
        return self._reverse(n, node)
```
