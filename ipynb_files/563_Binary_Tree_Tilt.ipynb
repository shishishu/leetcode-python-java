### Solution 1: Recursive

#### code

```python
class Solution:
    def findTilt(self, root):
        if not root:
            return 0
        self.left_vals = []
        self.right_vals = []
        if root.left:
            self.getValues(root.left, self.left_vals)
        if root.right:
            self.getValues(root.right, self.right_vals)
        return abs(sum(self.left_vals) - sum(self.right_vals)) + self.findTilt(root.left) + self.findTilt(root.right)
    
    
    def getValues(self, root, vals):
        if not root:
            return
        vals.append(root.val)
        if root.left:
            self.getValues(root.left, vals)
        if root.right:
            self.getValues(root.right, vals)
```

### Solution 2: Recursive2

#### code

```python
class Solution:
    def findTilt(self, root, ans=True):
        if not root:
            return 0 if ans else (0, 0)
        ls, ld = self.findTilt(root.left, False)
        rs, rd = self.findTilt(root.right, False)
        result = abs(ls - rs) + ld + rd
        if not ans:
            result = root.val + ls + rs, result
        return result 
```

### Solution 3:

#### code

```python
class Solution:
    
    res = 0
    
    def findTilt(self, root):
        self.recUnit(root)
        return res
    
    def recUnit(self, root):
        if not root:
            return 0
        left = recUnit(root.left)
        right = recUnit(root.right) 
        res += abs(left - right)
        return root.val + left + right
```
