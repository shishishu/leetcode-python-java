### Solution 1: ignore, issue occurs while big number

#### code

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
class Solution:
    def addTwoNumbers(self, l1, l2):
        n1 = self.convert_digit_to_number(self.extract_digit_from_listnode(l1))
        n2 = self.convert_digit_to_number(self.extract_digit_from_listnode(l2))
        n3 = n1 + n2
        return self.convert_digit_to_listnode(self.convert_number_to_digit(n3))
    
    def extract_digit_from_listnode(self, listnode):
        digit_list = []
        while (listnode):
            digit_list.append(listnode.val)
            listnode = listnode.next
        return digit_list
    
    def convert_digit_to_listnode(self, digit_list):
        listnode = ListNode(digit_list[0])
        headnode = listnode
        for digit in digit_list[1:]:
            tmpnode = ListNode(digit)
            listnode.next = tmpnode
            listnode = tmpnode
        return headnode
    
    def convert_digit_to_number(self, digit_list):
        number = 0
        for digit in digit_list[::-1]:
            number = 10 * number + digit
        return number
    
    def convert_number_to_digit(self, number):
        digit_list = []
        if number == 0:
            return [0]
        while (number > 0):
            digit_list.append(number % 10)
            number = int(number / 10)
        return digit_list
    
```

```python
def traverse_node(listnode):
    while listnode:
        print(listnode.val)
        listnode = listnode.next
```

#### test

```python
s = Solution()
s.convert_number_to_digit(0)
```

```python
l1 = s.convert_digit_to_listnode([2, 4, 3])
l2 = s.convert_digit_to_listnode([5, 6, 4])
```

```python
l3 = s.addTwoNumbers(l1, l2)
```

```python
traverse_node(l3)
```

### Solution 2

#### code

```python
class Solution:
    def addTwoNumbers(self, l1, l2):
        result = ListNode(0)
        result_tail = result
        carry = 0
        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0
            carry, out = divmod(val1 + val2 + carry, 10)
            
            result_tail.next = ListNode(out)
            result_tail = result_tail.next
            
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        return result.next
```
