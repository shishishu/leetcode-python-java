### Solution 1: inorder

#### code

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution:
    def getMinimumDifference(self, root):
        L = []
        def inorder(root):
            if root.left:
                inorder(root.left)
            L.append(root.val)
            if root.right:
                inorder(root.right)
        inorder(root)
        return min(b - a for a, b in zip(L[:-1], L[1:]))
```

### Solution 2: Recursive

#### code

```python
class Solution:
    def getMinimumDifference(self, root):
        def fn(node, lo, hi):
            if not node:
                return hi - lo
            left = fn(node.left, lo, node.val)
            right = fn(node.right, node.val, hi)
            return min(left, right)
        return fn(root, float('-inf'), float('inf'))
```
