### Solution 1: Brute Force (O(n3))

#### code

```python
class Solution:
    def validateStackSequences(self, pushed, popped):
        assert len(pushed) == len(popped)
        length = len(pushed)
        if length >= 3:
            for i in range(0, length - 2):
                idx_i = popped.index(pushed[i])
                for j in range(i + 1, length - 1):
                    idx_j = popped.index(pushed[j])
                    for k in range(j + 1, length):
                        idx_k = popped.index(pushed[k])
                        if idx_k < idx_i and idx_i < idx_j:
                            return False
        return True
```

```python
pushed = [1,2,3,4,5]
popped = [4,5,3,2,1]
```

```python
s = Solution()
s.validateStackSequences(pushed, popped)
```

### Solution 2: Brute Force (O(n2))

#### code

```python
class Solution:
    def validateStackSequences(self, pushed, popped):
        assert len(pushed) == len(popped)
        length = len(pushed)
        if length >= 3:
            for i in range(0, length - 2):
                idx_i = popped.index(pushed[i])
                for j in range(i + 1, length - 1):
                    idx_j = popped.index(pushed[j])
                    idx_k = popped.index(pushed[j + 1])
                    if idx_k < idx_i and idx_i < idx_j:
                        return False
        return True
```

### Solution 3: sequence check

#### code

```python
class Solution:
    def validateStackSequences(self, pushed, popped):
        inter_stack = []
        for elem in popped:
            while (not inter_stack or (inter_stack and inter_stack[-1] != elem)) and (pushed and pushed[0] != elem):
                inter_stack.append(pushed[0])
                pushed = pushed[1:]
            if pushed and pushed[0] == elem:
                pushed = pushed[1:]
                popped = popped[1:]
                continue
            if inter_stack and inter_stack[-1] == elem:
                inter_stack.pop()
                popped = popped[1:]
                continue
            return False
        return True
```

```python
pushed = [0,2,1,3]
popped = [1,2,3,0]
s = Solution()
s.validateStackSequences(pushed, popped)
```

### Solution 4: mock the process

#### code

```python
class Solution:
    def validateStackSequences(self, pushed, popped):
        idx = 0
        stack = []
        for p in pushed:
            stack.append(p)
            while stack and stack[-1] == popped[idx]:
                stack.pop()
                idx += 1
        return not stack
```

```python
pushed = [0,2,1,3]
popped = [1,2,3,0]
s = Solution()
s.validateStackSequences(pushed, popped)
```
