### Solution 1

#### code

```python
class Solution:
    def removeElement(self, nums, val):
        begin = 0
        for num in nums:
            if num != val:
                nums[begin] = num
                begin += 1
        return begin
```

#### test

```python
nums = [2,3,3,2]
val = 3
```

```python
s = Solution()
```

```python
s.removeElement(nums, val)
```
