### Solution 1: 

#### code

```python
d = {'M':1000, 'D':500, 'C':100, 'L':50, 'X':10, 'V':5, 'I':1}

class Solution:
    def romanToInt(self, s: str):
        res, p = 0, 'I'
        for c in s[::-1]:
            res, p = res - d[c] if d[c] < d[p] else res + d[c], c
        return res
```

```python
exp = "MCMXCIV"
```

```python
s = Solution()
```

### Solution 2: 

```python
d = {'M':1000, 'D':500, 'C':100, 'L':50, 'X':10, 'V':5, 'I':1}

class Solution:
    def romanToInt(self, s: str):
        res = 0
        res += d[s[-1]]
        for i in range(len(s)-2, -1, -1):
            res = res - d[s[i]] if d[s[i]] < d[s[i+1]] else res + d[s[i]]
        return res
```

```python
s = Solution()
```

```python
s.romanToInt("MCMXCIV")
```
