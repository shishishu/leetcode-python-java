### Solution 1: Brute Force

#### code

```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(0, len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
        return
```

#### test

```python
nums = [2, 7, 11, 15]
target = 9
```

```python
s = Solution()
s.twoSum(nums, target)
```

### Solution 2

#### code

```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(0, len(nums)-1):
            target_partner = target - nums[i]
            if target_partner in nums[i+1:]:
                return [i, nums[i+1:].index(target_partner)+i+1]
        return
```

#### test

```python
nums = [3,3]
target = 6
```

```python
s = Solution()
s.twoSum(nums, target)
```

### Solution 3: dict

#### code

```python
class Solution:
    def twoSum(self, nums, target):
        d = dict()
        for i, n in enumerate(nums):
            m = target - n
            if m in d:
                return [i, d[m]]
            else:
                d[n] = i
        return 
```
