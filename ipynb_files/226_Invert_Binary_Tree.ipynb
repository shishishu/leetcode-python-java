### Solution 1: default

#### code

```python
class Solution:
    def invertTree(self, root):
        def swapLR(root):
            if root:
                tmp = root.left
                root.left = root.right
                root.right = tmp
                swapLR(root.left)
                swapLR(root.right)
        swapLR(root)
        return root
```

### Solution 2: Recursive

#### code

```python
class Solution:
    def invertTree(self, root):
        if root:
            root.left, root.right = self.invertTree(root.right), self.invertTree(root.left)
        return root
```

### Solution 3: Iterative

#### code

```python
class Solution:
    def invertTree(self, root):
        stack = [root]
        while stack:
            curr = stack.pop()
            if curr:
                curr.left, curr.right = curr.right, curr.left
                stack.append(curr.left)
                stack.append(curr.right)
        return root
```
