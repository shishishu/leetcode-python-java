### Solution 1: bubble sort

#### code

```python
class Solution:
    def sortArray(self, nums):
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] > nums[j]:
                    nums[i], nums[j] = nums[j], nums[i]
        return nums
```

#### test

```python
nums = [5,2,3,1]
```

```python
s = Solution()
s.sortArray(nums)
```

```python
s.sortArray([5,1,1,2,0,0])
```

### Solution 2: bucket sort

#### code

```python
class Solution:
    def sortArray(self, nums):
        shift_cnt = 50000
        cnt_list = [0] * 100001
        idx = 0
        for i in range(len(nums)):
            cnt_list[nums[i] + 50000] += 1
        for i in range(len(cnt_list)):
            while(cnt_list[i] > 0):
                nums[idx] = i - 50000
                idx += 1
                cnt_list[i] -= 1
        return nums
```
