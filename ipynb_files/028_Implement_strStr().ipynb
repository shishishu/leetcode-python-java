### Solution 1: Brute Force

#### code

```python
class Solution:
    def strStr(self, haystack, needle):
        n = len(haystack)
        m = len(needle)
        i = 0
        j = 0
        if m == 0:
            return 0
        while (i < n) and (j < m):
            if haystack[i] == needle[j]:
                i += 1
                j += 1
            else:
                i = i - j + 1
                j = 0
        return (i - j) if (i - j <= n - m) else -1
```

#### test

```python
s = Solution()
```

```python
haystack = "hello"
needle = "ll"
s.strStr(haystack,  needle)
```

### Solution 2: Brute Force

#### code

```python
class Solution:
    def strStr(self, haystack, needle):
        n = len(haystack)
        m = len(needle)
        i = 0
        j = 0
        if m == 0:
            return 0
        for i in range(n - m + 1):
            for j in range(m):
                if haystack[i + j] != needle[j]:
                    j -= 1
                    break
            if (j >= m - 1):
                break
        return i if j == m - 1 and n >= m else -1
```

### Solution 3: KMP

#### code

```python
class Solution:
    def createNext(self, p):
        t = -1
        m = len(p)
        nextArr = [-2] * m
        nextArr[0] = t
        j = 0
        while j < m - 1:
            if (t < 0) or (p[j] == p[t]):
                t += 1
                j += 1
                nextArr[j] = t
            else:
                t = nextArr[t]
        return nextArr
    
    def strStr(self, haystack, needle):
        n = len(haystack)
        m = len(needle)
        if m == 0:
            return 0
        nextArr = self.createNext(needle)
        i = j = 0
        while (i < n) and (j < m):
            if haystack[i] == needle[j]:
                i += 1
                j += 1
            else:
                if j == 0:
                    i += 1
                else:
                    j = nextArr[j]
        return (i - j) if (j == m) else -1
```
