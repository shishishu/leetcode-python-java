### Solution 1: inserted sort func

#### code

```python
class Solution:
    def merge(self, intervals):
        if len(intervals) <= 1:
            return intervals
        out = []
        for interval in sorted(intervals):
            if not out:
                out.append(interval)
            else:
                if interval[0] <= out[-1][1]:
                    out[-1][1] = max(interval[1], out[-1][1])
                else:
                    out.append(interval)
        return out
```
