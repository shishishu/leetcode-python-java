### Solution 1: 

#### code

```python
values = [ 1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1 ]
numerals = [ "M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I" ]

class Solution:
    def intToRoman(self, num: int):
        res = ""
        i = 0
        while(num != 0):
            while (num >= values[i]):
                num -= values[i]
                res += numerals[i]
            i += 1
        return res
```

#### test

```python
s = Solution()
```

```python
s.intToRoman(1994)
```

### Solution 2: 

```python
values = [ 1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1 ]
numerals = [ "M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I" ]

class Solution:
    def intToRoman(self, num: int):
        res = ""
        for i, v in enumerate(values):
            res += (num // v) * numerals[i]
            num %= v
        return res
```

```python
s = Solution()
```

```python
s.intToRoman(1994)
```
