### Solution 1: Brute Force

#### code

```python
class Solution:
    def containsNearbyAlmostDuplicate(self, nums, k, t):
        length = len(nums)
        for i in range(length - 1):
            for j in range(i + 1, min(length, i + k + 1)):
                if abs(nums[i] - nums[j]) <= t:
                    return True
        return False
```

#### test

```python
nums = [1,2,3,1]
k = 3
t = 0
```

```python
s = Solution()
```

```python
s.containsNearbyAlmostDuplicate(nums, k, t)
```

### Solution 2: Use Set

#### code

```python
class Solution:
    def containsNearbyAlmostDuplicate(self, nums, k, t):
        length = len(nums)
        a = set()
        for i in range(length):
            if t == 0:
                if nums[i] in a:
                    return True
            else:
                for item in a:
                    if abs(nums[i] - item) <= t:
                        return True
            a.add(nums[i])
            if len(a) == k + 1:
                a.remove(nums[i-k])
        return False
```

### Solution 3: Bucket

#### code

```python
class Solution:
    def containsNearbyAlmostDuplicate(self, nums, k, t):
        if t < 0:
            return False
        w = t + 1 
        d = dict()
        for i in range(len(nums)):
            m = nums[i] // w  # bucket index
            if m in d:
                return True
            elif (m-1 in d and abs(d[m-1] - nums[i]) <= t):
                return True
            elif (m+1 in d and abs(d[m+1] - nums[i]) <= t):
                return True
            d[m] = nums[i]
            if (i >= k):
                d.pop(nums[i-k] // w)
        return False
```
