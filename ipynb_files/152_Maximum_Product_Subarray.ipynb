### Solution 1: Brute Force

#### code

```python
class Solution:
    def maxProduct(self, nums):
        if not nums:
            return None
        max_val = nums[0]
        length = len(nums)
        for i in range(length):
            output = 1
            for j in range(i, length):
                output *= nums[j]
                if output > max_val:
                    max_val = output
        return max_val
```

#### test

```python
nums = [-2,0,-1]
```

```python
s = Solution()
```

```python
s.maxProduct(nums)
```

### Solution 2: Dynamic Programming

#### code

```python
class Solution:
    def maxProduct(self, nums):
        r = nums[0]
        imax = r
        imin = r
        for i in range(1, len(nums)):
            if nums[i] < 0:
                imax, imin = imin, imax
            imax = max(nums[i], nums[i] * imax)
            imin = min(nums[i], nums[i] * imin)
            r = max(r, imax)
        return r
```

#### test

```python
s = Solution()
```

```python
s.maxProduct(nums)
```
