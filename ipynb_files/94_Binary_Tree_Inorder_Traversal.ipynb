### Solution 1: Recursive

#### code

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution:
    def inorderTraversal(self, root):
        output = []
        self.rec_inorder(root, output)
        return output
    
    def rec_inorder(self, root, output):
        if root:
            self.rec_inorder(root.left, output)
            output.append(root.val)
            self.rec_inorder(root.right, output)
```

```python
import math
def generate_tree(tree_vals):
    num_elem = len(tree_vals)
    num_level = math.log2(num_elem + 1)
    assert num_level % 1 == 0, 'wrong number in tree values...'
    # init all tree nodes
    ptree = []
    for i in range(num_elem):
        tmp_tree = TreeNode(tree_vals[i])
        ptree.append(tmp_tree)
    for i in range(int(num_elem / 2 - 1) + 1):
        ptree[i].left = ptree[2 * i + 1]
        ptree[i].right = ptree[2 * i + 2]
    return ptree[0]
```

#### test

```python
tree_vals = [1, 2, 3, 4, 5, 6, 7]
root = generate_tree(tree_vals)
```

```python
s = Solution()
```

```python
s.inorderTraversal(root)
```

### Solution 3: Iterative (general)

#### code

```python
class Solution:
    def inorderTraversal(self, root):
        output = []
        stack = []
        while True:
            self.goAlongLeftBranch(root, stack)
            if not stack:
                break
            root = stack.pop()
            output.append(root.val)
            root = root.right
        return output
    
    def goAlongLeftBranch(self, root, stack):
        while root:
            stack.append(root)
            root = root.left
```
