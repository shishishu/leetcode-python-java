### Solution 1: Iterative

#### code

```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        if not s:
            return 0
        sub_list = []
        len_list = []
        sub_list.append(s[0])
        len_list.append(1)
        for idx in range(1, len(s)):
            if s[idx] not in sub_list[idx - 1]:
                elem = sub_list[idx - 1] + s[idx]
            else:
                if s[idx] == sub_list[idx - 1][-1]:
                    elem = s[idx]
                else:
                    elem = sub_list[idx - 1][(sub_list[idx - 1].index(s[idx]) + 1):] + s[idx]
            sub_list.append(elem)
            len_list.append(len(elem))
        return max(len_list)
```

### Solution 2: Sliding window

#### code

```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        n = len(s);
        ans = i = j = 0
        charSet = set()
        while (i < n and j < n):
            if not s[j] in charSet:
                charSet.add(s[j])
                j += 1
                ans = max(ans, j - i)
            else:
                charSet.remove(s[i])
                i += 1
        return ans
```
