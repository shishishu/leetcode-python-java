### Solution 1: Recursive

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

#### code

```python
class Solution:
    def generateTrees(self, n):
        def node(val, left, right):
            node = TreeNode(val)
            node.left = left
            node.right = right
            return node
        def trees(lo, hi):
            return [node(root, left, right) for root in range(lo, hi+1) for left in trees(lo, root - 1) for right in trees(root+1, hi)] or [None]
        if n == 0:
            return []
        return trees(1, n)
```

### Solution 2: DP

#### code

```python
class Solution:
    def generateTrees(self, n):
        res = [[] for i in range(n+1)]
        if n == 0:
            return res[0]
        res[0].append(None)
        res[1].append(TreeNode(1))
        for i in range(2, n + 1):
            for j in range(1, i+1):
                for nodeL in res[j - 1]:
                    for nodeR in res[i - j]:
                        node = TreeNode(j)
                        node.left = nodeL
                        node.right = self.cloneTrees(nodeR, j)
                        res[i].append(node)
        return res[n]
        
    def cloneTrees(self, node, offset):
        if not node:
            return None
        newNode = TreeNode(node.val + offset)
        newNode.left = self.cloneTrees(node.left, offset) 
        newNode.right = self.cloneTrees(node.right, offset)
        return newNode
```
