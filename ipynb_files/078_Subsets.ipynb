### Solution 1

#### code

```python
class Solution:
    def subsets(self, nums):
        output = [[]]
        for num in nums:
            output += [curr + [num] for curr in output]
        return output
```

#### test

```python
s = Solution()
```

```python
s.subsets([1,2,3])
```
